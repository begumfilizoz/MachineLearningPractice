```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv')
```

```python
#Showing the first few items in the dataset
```

```python
df.head()
```

```python
#Removing unnecessary columns
```

```python
df = df.drop(columns='id', axis = 1)
```

```python
df = df.drop(columns='Unnamed: 32', axis = 1)
```

```python
df.head()
```

```python
#Labeling malignant cells as 1, benign ones as 0
```

```python
df['diagnosis'] = df['diagnosis'].map({'M' : 1, 'B' : 0})
```

```python
df.head()
```

```python
df.columns
```

```python
mean_features = list(df.columns[1:11])
se_features = list(df.columns[11:21])
worst_features = list(df.columns[21:31])
```

```python
print(se_features)
```

```python
mean_features.append('diagnosis')
se_features.append('diagnosis')
worst_features.append('diagnosis')
```

```python
corr = df[worst_features].corr()
```

```python
corr
```

```python
#Training the model
```

```python
prediction_vars = ['radius_mean', 'perimeter_mean', 'area_mean', 'compactness_mean', 'concavity_mean', 'concave points_mean', 'radius_se', 'area_se', 'radius_worst', 'perimeter_worst', 'compactness_worst']
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(df, test_size = 0.15, random_state = 1)
```

```python
train_x = train[prediction_vars]
train_y = train['diagnosis']
test_x = test[prediction_vars]
test_y = test['diagnosis']
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

```python
model = RandomForestClassifier()
```

```python
model.fit(train_x, train_y)
```

```python
predictions = model.predict(test_x)
```

```python
test_y
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(test_y, predictions)
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
```

```python
#Evaluating our results
```

```python
precision = precision_score(test_y, predictions)
```

```python
print("This is the precision of the predictions: %.3f" % precision)
```

```python
recall = recall_score(test_y, predictions)
```

```python
print("This is the recall of the predictions: %.3f" % recall)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(test_y, predictions)
```

```python
print("This is the accuracy of the predictions: %.3f" % accuracy)
```
